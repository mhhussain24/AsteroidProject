```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
miss_df_to_load = "neo_v2.csv" 
```

```python
miss_df = pd.read_csv(miss_df_to_load)
miss_df.head(20)
```

```python
miss_df['year'] = miss_df['name'].str.extract('(\(\d\d\d\d)', expand=True)
miss_df['year'] = miss_df['year'].str[1:5]
```

```python
miss_df.head(20)
```

```python
miss_df['year'] = miss_df['name'].apply(lambda x: x.split(' ')[0][1:])
miss_df.head()
```

```python
hazardous_true = miss_df.loc[miss_df['hazardous'] == True]
hazardous_true
```

```python
hazardous_false = miss_df.loc[miss_df['hazardous'] == False]
hazardous_false
```

```python
hazardous_count_false = hazardous_false.groupby(["id", "hazardous"])
hazardous_count_false.head()
```

```python
# Build the scatter plot for latitude vs. max temperature.
plt.scatter(miss_df.loc[0:500, "relative_velocity"],
            miss_df.loc[0:500, "miss_distance"],
            edgecolor="black", linewidths=1, marker="o",
            alpha=0.8, label="Cities")

# Incorporate the other graph properties.
plt.title(f"Miss Distance vs Relative Velocity")
plt.ylabel("relative_velocity")
plt.xlabel("miss_distance")
plt.grid(True)

# Save the figure.
#plt.savefig("velocity_distance/Fig1.png")

# Show plot.
plt.show()
```
