```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
miss_df_to_load = "neo_v2.csv" 
```

```python
miss_df = pd.read_csv(miss_df_to_load)
miss_df.head()
```

```python
miss_df['year'] = miss_df['name'].apply(lambda x: x.split(' ')[0][1:])
miss_df.head()
```

```python
# Create a box-and-whisker plot for hazardous count.
x_labels = ["est_diameter_min"]
fig, ax = plt.subplots()
ax.boxplot(hazardous_true["est_diameter_min"], labels=x_labels)
# Add the title, y-axis label and grid.
ax.set_title('Hazardous Asteroids')
ax.set_ylabel('Number of Asteroids')
ax.set_yticks(np.arange(10, 41, step=2.0))
ax.grid()
plt.show()
```
